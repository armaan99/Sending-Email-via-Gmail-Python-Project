# Sending Email using Python

```python
import email            # pip install email
import smtplib          # pip install smtplib
import ssl              # pip install ssl
import base64           # pip install pybase64
import os               # pip install os-sys
import pathlib          # pip install pathlib


# Multipart: for connecting multiple content parts to form a message, the parts can be different types of data.
from email.mime.multipart import MIMEMultipart

'''Types of Media and their subtypes'''
# Text: for standard representation of a text message, which may consist of various character sets or formats
    # Available Options : plain, html
# Image: for transfer of static image data.
    # Available Options : gif, jpeg, png
# Audio: for transfer of audio or sound data.
# Video: for transfer of dynamic image data, which may be a video data format that includes audio.
    # Available Options : mpeg
# Application: for transfer of application data or binary data.
    # Available Options : xhtml+xml, octet-stream
# Message: for packing an email message.
    # Available Options : rfc822
from email.mime.application import MIMEApplication
from email.mime.text import MIMEText
from email import encoders
```

```python
class MailAssistant():

    
    sender_email = ""
    sender_password = ""
    reciever_emails = []
    body = ""
    subject = ""
    

    def __setSenderCredentials(self):
        '''
        Privately Accessible method for
        Taking input as Sender's Email ID and Password
        '''
        self.sender_email = input("Enter your Email ID : ").lower()
        self.sender_password = input("Enter your Password (Case-sensitive) : ")


    def __setRecieverDetails(self):
        '''
        Privately Accessible method for
        Taking input as number of recievers and their Mail IDs
        '''
        for x in range(int(input("To how many people you want to send this E-mail : "))):
            self.reciever_emails.append(input("Enter E-mail ID of reciever {} : ".format(x+1)))

    
    def __setMessage(self):
        '''
        Privately Accessible method for
        Taking Message Subject and Mail content as an input
        '''
        self.subject = input("Type your subject here : ")       # Subject as the input
        self.body = input("Type your message here : ")          # Main Content as the input


    def __addAttachments(self):
        '''
        Privately Accessible method for
        Attaching files to mail
        '''
        path = input("Enter full file path(with name and extension) : ")
        if input("Do you want to attach file with another name?(y/n) : ").lower() == 'y':
            fileName = input("Enter new file Name : ") + pathlib.Path(path).suffix
        else:
            fileName = os.path.basename(path)
        File = MIMEApplication(open(path, 'rb').read())
        File.add_header('Content-Disposition', 'attachment', filename=fileName)
        return File
        
    
    def __emailSkeleton(self):
        '''
        Privately Accessible method for
        Creating Email Skeleton(for Gmail)
        '''
        # Gmail takes Reciever's Email-IDs as string which consists of Mail Ids separated by and ending with comma(,)
        email_ids = ",".join(self.reciever_emails) + ","
        
        # Message Skeleton begins from here
        message = MIMEMultipart()

        message['From'] = self.sender_email                     # Sender
        message['To'] = email_ids                               # Reciever(s)
        message['Subject'] = self.subject                       # Email Subject
        message['Bcc'] = email_ids                              # BCC Email IDs (in this case, same as recievers)
        
        message.attach(MIMEText(self.body, "plain"))            # Main Content/Body of mail

        while input("Do you want to attach any file(y/n) : ").lower() == 'y':
            message.attach(self.__addAttachments())             # File attachments
        
        return message.as_string()


    def sendMail(self):
        '''
        Publicly Accessible method for
        Sending Mail
        '''
        # Taking necessary feilds of mail
        self.__setSenderCredentials()
        self.__setRecieverDetails()
        self.__setMessage()

        # Framing message skeleton
        message = self.__emailSkeleton()

        # Establishing Connection with Gmail
        server = smtplib.SMTP("smtp.gmail.com",587)     # Creating SMTP Session; parameters-(Host, Port)
        server.ehlo()                                   # Must command for the SMPT client to commence a mail transfer
        server.starttls()                               # For security
        
        # Sending Mail
        if self.sender_email != "" and self.sender_password != "" and self.reciever_emails != [] and message != "":
            server.login(self.sender_email, self.sender_password)
            server.sendmail(self.sender_email, self.reciever_emails, message)
            print("Your E-mail sent successfully")
        else:
            print("Details are insufficient")

        # Closing Connection
        server.close()
```

```python
mailAssistant = MailAssistant()
mailAssistant.sendMail()
```
